# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

## Подготовка данных

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.utils import shuffle


from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_curve, roc_auc_score, confusion_matrix

from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
 

data = pd.read_csv('/datasets/Churn.csv')
data.info()
```

Имеем пропуски в Tenure

```python
data.isnull().sum()
```

```python
data.describe()
```

Пропущенные значения в столбце `Tenure` заменим на 0, так как есть предположение, что эти клиенты пользуются меньше года


```python
data['Tenure'] = data['Tenure'].replace(np.nan, 0)
```

```python
data.info()
```

Присутсвует дисбаланс классов. Людей, которые не собираются уходить около 80%

```python
data['Exited'].value_counts(normalize=True)
```

Balance, EstimatedSalary округлим вверх и приведём к типу int

```python
data['Balance'] = data['Balance'].astype(int)
data['EstimatedSalary'] = data['EstimatedSalary'].astype(int)
data['Tenure'] = data['Tenure'].astype(int)

data.head()
```

```python
data.info()
```

Обучим модель без учёта дисбаланса классов

```python
data['Geography'].unique()
```

Сформируем таблицу с признаками, удалив ненужные.Далее создадим дамми-переменные с 3мя странами, указав праметр drop_first, чтобы избежать ловушки. 

```python
# features = pd.get_dummies(data=data, columns=['Geography'], drop_first=True)
data = data.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
```

```python
data.info()
```

```python
data = pd.get_dummies(data=data, columns=['Geography', 'Gender'], drop_first=True)
data
```

```python
data.info()
```

Данные приведены в порядок, а теперь можно и разбить исходную выборку на train, valid, test

```python
target = data['Exited']
features = data.drop(columns=['Exited'])

features_train, features_valid, target_train, target_valid = train_test_split( \
    features, target, test_size=.40, random_state=12345) 

features_valid, features_test, target_valid, target_test = train_test_split( \
    features_valid, target_valid, test_size=.50, random_state=12345) 
```

```python
features.info()
```

```python
#Приведём данные к общему масштабу по числовым признакам

numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
scaler = StandardScaler()

features_train.loc[:,numeric] = scaler.fit_transform(features_train[numeric])

features_train.head(15)
```


Такая команда (можно поместить в самое начало тетрадки) иногда может помочь не выводить предупреждения)<br>
import warnings
warnings.filterwarnings('ignore')

```python
features_valid.loc[:, numeric]  = scaler.transform(features_valid[numeric])
features_valid
```

```python
features_test.loc[:, numeric]  = scaler.transform(features_test[numeric])
features_test
```

```python
data.info()
```

# Вывод (Шаг 1)
Предобработали датасет, а именно занулили пропуски в столбце `Tenure`, убрали ненужные для модели колонки. Далее разбили датасет на выборки train, valid, test в соотношении 3:2:2. Потом стандартизировали числовые признаки.

## Исследование задачи

Посмотрим на работу наших моделей без учёта дисбаланса классов.

```python


# Логистическая регрессия без учёта дисбаланса
model = LogisticRegression(random_state=12345, solver='liblinear')
model.fit(features_train, target_train)
pred_valid = model.predict(features_valid)
print('Accuracy:', accuracy_score(pred_valid, target_valid))
print('Confusion Matrix:', confusion_matrix(pred_valid, target_valid))
print('Precision:', precision_score(pred_valid, target_valid))
print('Recall:', recall_score(pred_valid, target_valid))
print('F1:', f1_score(pred_valid, target_valid))
```

```python
#Решающее дерево


best_model = None
conf_matrix = None
best_score = 0

for d in range(1, 51):
    model = DecisionTreeClassifier(random_state=12345, max_depth=d)
    model.fit(features_train, target_train)
    pred_valid = model.predict(features_valid)
    if accuracy_score(pred_valid, target_valid) > best_score:
        best_score = accuracy_score(pred_valid, target_valid)
        conf_matrix = confusion_matrix(pred_valid, target_valid)
        best_model = model

print('Depth: ', best_model.max_depth)    
print('Accuracy: ', best_score)
print('Confusion Matrix:', conf_matrix)
print('Precision:', precision_score(pred_valid, target_valid))
print('Recall:', recall_score(pred_valid, target_valid))
print('F1:', f1_score(pred_valid, target_valid))
    
    
    
```

```python
# Случайный лес

best_model = None
conf_matrix = None
best_score = 0


for d in range(1, 100):
    model = RandomForestClassifier(random_state=12345, max_depth=d, n_estimators=120)
    model.fit(features_train, target_train)
    pred_valid = model.predict(features_valid)
    if accuracy_score(pred_valid, target_valid) > best_score:
        best_score = accuracy_score(pred_valid, target_valid)
        conf_matrix = confusion_matrix(pred_valid, target_valid)
        best_model = model

print('Depth: ', best_model.max_depth)
print('n_estimators: ', best_model.n_estimators)
print('Accuracy: ', best_score)
print('Confusion Matrix:', conf_matrix)
print('Precision:', precision_score(pred_valid, target_valid))
print('Recall:', recall_score(pred_valid, target_valid))
print('F1:', f1_score(pred_valid, target_valid))
    
```

Видно, что чуть лучше справляестя модель случайного леса, так как у неё меньше ложно-положительных (FP) предсказаний и метрика F1 превосходит своих конкурентов.

# Вывод (Шаг 2)

 В генеральной совокупности имеем дисбаланс классов в соотношении 4:1, где меньшая доля - единицы в столбце Exited, означающие , что клиент покинул банк. Т.к имеем дисбаланс метрика Accuracy теряет свою альтернативность, т.к если её сравнивать с моделью которая предсказывает одни нули - accuracy был бы высок и совпадал бы с соотношением дисбаланса.
 При дисбалансе классов метрика Precision не превышала значение 0,5, а это значит, что модели при дисбалансе классов предсказали меньше половины отрицательных значений. 

## Борьба с дисбалансом

Попробуем обучить модель логисчтической регрессии с парметром class_weight=balanced, который присваивает больший вес малому классу.

```python
# Функция для отрисовки ROC-curve
def plot_roc_curve(fpr, tpr, thres):
    plt.figure()

    plt.plot(fpr, tpr)

    plt.plot([0, 1], [0, 1], linestyle='--')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show()
```

```python
model = LogisticRegression(random_state=12345, solver='liblinear', class_weight='balanced')
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)

probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid)

plot_roc_curve(fpr, tpr, thresholds)

print('Accuracy:', accuracy_score(predicted_valid, target_valid))
print('Precision:', precision_score(predicted_valid, target_valid))
print('Recall:', recall_score(predicted_valid, target_valid))
print('F1:', f1_score(predicted_valid, target_valid))
print('ROC-AUC', roc_auc_score(target_valid, probabilities_one_valid))

print('Confusion Matrix:', confusion_matrix(predicted_valid, target_valid))
# print(thresholds)
```

```python

probs = model.predict_proba(features_valid)

# Вероятности принадлежности к перовму классу
probs_one_valid = probs[:, 1]

# Поробуем изменить порог для классификации

for threshold in np.arange(0, 0.7, 0.02):
    predicted_valid = probs_one_valid > threshold  # < напишите код здесь >
    precision = precision_score(target_valid, predicted_valid) # < напишите код здесь >
    recall = recall_score(target_valid, predicted_valid) # < напишите код здесь >

    print("Порог = {:.2f} | Точность = {:.3f}, Полнота = {:.3f}".format(
        threshold, precision, recall))

```

```python
#Попробуем применить upsampling для нашей выборки и обучить модели
features_zeros = features_train[target_train == 0]
features_ones = features_train[target_train == 1]
target_zeros = target_train.loc[target_train == 0]
target_ones = target_train.loc[target_train == 1]

features_upsampled = pd.concat([features_zeros] + [features_ones] * 4)
target_upsampled = pd.concat([target_zeros] + [target_ones] * 4)

```

```python
# После upsampling восстановили баланс, а теперь перемешаем наши данные, чтобы отдать модели
target_upsampled.value_counts(normalize=True)
features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)

```

```python
# Обучим на новых данных логистическую регрессию

model = LogisticRegression(random_state=12345, solver='liblinear')
model.fit(features_upsampled, target_upsampled)
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid)

plot_roc_curve(fpr, tpr, thresholds)


predicted_valid = model.predict(features_valid)

print("F1:", f1_score(target_valid, predicted_valid))
print("Precision:", precision_score(target_valid, predicted_valid))
print("Accuracy:", accuracy_score(target_valid, predicted_valid))
print('ROC-AUC', roc_auc_score(target_valid, probabilities_one_valid))



```


что-то логистическая - не боец в этом проекте)

```python
# Попробуем бывалую модель случайного леса

model = RandomForestClassifier(random_state=12345, max_depth=13, n_estimators=120)
model.fit(features_upsampled, target_upsampled)
pred_valid = model.predict(features_valid)

print("F1:", f1_score(target_valid, pred_valid))
print("Precision:", precision_score(target_valid, pred_valid))
print("Accuracy:", accuracy_score(target_valid, pred_valid))
```

F1 метрика чуть преуспела, по сравнению с дисбалансированной выборкой. Попробуем переобучить модель.

```python
best_model = None
conf_matrix = None
best_score = 0


for d in range(10, 20):
    for n in range(100, 150):
        model = RandomForestClassifier(random_state=12345, max_depth=d, n_estimators=n)
        model.fit(features_upsampled, target_upsampled)
        pred_valid = model.predict(features_valid)
        if f1_score(pred_valid, target_valid) > best_score:
            best_score = f1_score(pred_valid, target_valid)
            conf_matrix = confusion_matrix(pred_valid, target_valid)
            best_model = model

print('Depth: ', best_model.max_depth)
print('n_estimators: ', best_model.n_estimators)
print('Accuracy: ', best_score)
print('Confusion Matrix:', conf_matrix)
print('Precision:', precision_score(pred_valid, target_valid))
print('Recall:', recall_score(pred_valid, target_valid))
print('F1:', f1_score(pred_valid, target_valid))
    
```
